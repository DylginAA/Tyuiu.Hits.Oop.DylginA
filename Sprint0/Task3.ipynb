Описание задачи:
Создать базовый класс Inventory в C#, который будет представлять информацию о
наличии товаров на складе. На основе этого класса разработать 2-3 производных
класса, демонстрирующих принципы наследования и полиморфизма. В каждом из
классов должны быть реализованы новые атрибуты и методы, а также
переопределены некоторые методы базового класса для демонстрации
полиморфизма.

Требования к базовому классу Inventory:
• Атрибуты: 

ID склада (WarehouseId), 

Название склада (WarehouseName),


Общий объем хранения (StorageCapacity).




• Методы:



o GetStorageStatus(): метод для получения статуса доступного
пространства на складе.


o AddItem(Item item): метод для добавления товара на склад.


o RemoveItem(Item item): метод для удаления товара со склада.


Требования к производным классам:


1. ПерсональныйСклад (PersonalInventory): Должен содержать
дополнительные атрибуты, такие как Владелец склада (OwnerName).
Метод GetStorageStatus() должен быть переопределен для отображения
информации о владельце склада вместе с статусом хранения.
2. ГрупповойСклад (GroupInventory): Должен содержать дополнительные
атрибуты, такие как Группа товаров (ProductGroup). Метод AddItem() должен
быть переопределен для добавления информации о группе товаров при
добавлении нового товара.
3. АвтоматизированныйСклад (AutomatedInventory) (если требуется третий
класс): Должен содержать дополнительные атрибуты, такие как
Автоматизация уровня (AutomationLevel). Метод RemoveItem() должен быть
переопределен для добавления информации о уровне автоматизации при
удалении товара.

```
public class Inventory
{
    public string Name { get; set; }
    public double Capacity { get; set; }
    protected double UsedSpace { get; set; }
    protected List<string> Items { get; set; }

    public Inventory(string name, double capacity)
    {
        Name = name;
        Capacity = capacity;
        UsedSpace = 0;
        Items = new List<string>();
    }

    public virtual void ShowStatus()
    {
        Console.WriteLine($"{Name}: занято {UsedSpace}/{Capacity} м³");
    }

    public virtual void AddItem(string item, double size)
    {
        if (UsedSpace + size <= Capacity)
        {
            Items.Add(item);
            UsedSpace += size;
            Console.WriteLine($"Добавлен '{item}' на {Name}");
        }
        else
        {
            Console.WriteLine($"Не хватает места для '{item}' на {Name}");
        }
    }

    public virtual void RemoveItem(string item, double size)
    {
        if (Items.Remove(item))
        {
            UsedSpace -= size;
            Console.WriteLine($"Удален '{item}' с {Name}");
        }
        else
        {
            Console.WriteLine($"'{item}' не найден на {Name}");
        }
    }
}

// Склад 1
public class Inventory1 : Inventory
{
    public string Owner { get; set; }

    public Inventory1(double capacity, string owner) 
        : base("Склад 1", capacity)
    {
        Owner = owner;
    }

    public override void ShowStatus()
    {
        Console.WriteLine($"{Name} (Владелец: {Owner}): занято {UsedSpace}/{Capacity} м³");
    }
}

// Склад 2
public class Inventory2 : Inventory
{
    public string Category { get; set; }

    public Inventory2(double capacity, string category) 
        : base("Склад 2", capacity)
    {
        Category = category;
    }

    public override void AddItem(string item, double size)
    {
        if (UsedSpace + size <= Capacity)
        {
            Items.Add(item);
            UsedSpace += size;
            Console.WriteLine($"Добавлен '{item}' ({Category}) на {Name}");
        }
        else
        {
            Console.WriteLine($"Не хватает места для '{item}' на {Name}");
        }
    }
}

// Склад 3
public class Inventory3 : Inventory
{
    public int Level { get; set; }

    public Inventory3(double capacity, int level) 
        : base("Склад 3", capacity)
    {
        Level = level;
    }

    public override void RemoveItem(string item, double size)
    {
        if (Items.Remove(item))
        {
            UsedSpace -= size;
            Console.WriteLine($"Удален '{item}' (Уровень: {Level}) с {Name}");
        }
        else
        {
            Console.WriteLine($"'{item}' не найден на {Name}");
        }
    }
}

        Console.WriteLine("=== Система складов ===\n");

        Inventory1 sklad1 = new Inventory1(100, "Иванов");
        Inventory2 sklad2 = new Inventory2(80, "Электроника");
        Inventory3 sklad3 = new Inventory3(120, 3);

        Console.WriteLine("1. Добавление предметов:");
        sklad1.AddItem("Стол", 10);
        sklad1.AddItem("Стул", 5);
        
        sklad2.AddItem("Ноутбук", 3);
        sklad2.AddItem("Телефон", 1);
        
        sklad3.AddItem("Принтер", 8);
        sklad3.AddItem("Монитор", 12);

        Console.WriteLine("\n2. Статус складов:");
        sklad1.ShowStatus();
        sklad2.ShowStatus();
        sklad3.ShowStatus();

        Console.WriteLine("\n3. Удаление предметов:");
        sklad1.RemoveItem("Стул", 5);
        sklad3.RemoveItem("Принтер", 8);

        Console.WriteLine("\n4. статус:");
        sklad1.ShowStatus();
        sklad2.ShowStatus();
        sklad3.ShowStatus();
```

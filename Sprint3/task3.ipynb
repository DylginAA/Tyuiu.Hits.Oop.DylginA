```

public class Item
{
    public int ItemId { get; set; }
    public string Name { get; set; }
    public double Volume { get; set; } 

    public Item(int id, string name, double volume)
    {
        ItemId = id;
        Name = name;
        Volume = volume;
    }

    public override string ToString()
    {
        return $"[{ItemId}] {Name} (объем: {Volume} м³)";
    }
}

public class Inventory
{
    public int WarehouseId { get; set; }
    public string WarehouseName { get; set; }
    public double StorageCapacity { get; set; }
    protected List<Item> Items;

    public Inventory(int id, string name, double capacity)
    {
        WarehouseId = id;
        WarehouseName = name;
        StorageCapacity = capacity;
        Items = new List<Item>();
    }

    public virtual void AddItem(Item item)
    {
        double used = GetUsedVolume();
        if (used + item.Volume <= StorageCapacity)
        {
            Items.Add(item);
            Console.WriteLine($"Товар {item.Name} добавлен на склад {WarehouseName}");
        }
        else
        {
            Console.WriteLine("Недостаточно места на складе!");
        }
    }

    public void AddItem(Item item, int count)
    {
        for (int i = 0; i < count; i++)
            AddItem(item);
    }

    public virtual void RemoveItem(Item item)
    {
        if (Items.Remove(item))
            Console.WriteLine($"Товар {item.Name} удалён со склада {WarehouseName}");
        else
            Console.WriteLine("Товар не найден на складе!");
    }

    public virtual void GetStorageStatus()
    {
        Console.WriteLine($"Склад: {WarehouseName}. Занято {GetUsedVolume()} / {StorageCapacity} м³");
    }

    protected double GetUsedVolume()
    {
        double total = 0;
        foreach (var item in Items)
            total += item.Volume;
        return total;
    }
}

public class PersonalInventory : Inventory
{
    public string OwnerName { get; set; }

    public PersonalInventory(int id, string name, double capacity, string owner)
        : base(id, name, capacity)
    {
        OwnerName = owner;
    }

    public override void GetStorageStatus()
    {
        Console.WriteLine($"Персональный склад владельца {OwnerName}: занято {GetUsedVolume()} / {StorageCapacity} м³");
    }
}

public class GroupInventory : Inventory
{
    public string ProductGroup { get; set; }

    public GroupInventory(int id, string name, double capacity, string group)
        : base(id, name, capacity)
    {
        ProductGroup = group;
    }

    public override void AddItem(Item item)
    {
        Console.WriteLine($"Добавление товара в группу {ProductGroup}");
        base.AddItem(item);
    }
}

public class AutomatedInventory : Inventory
{
    public int AutomationLevel { get; set; }

    public AutomatedInventory(int id, string name, double capacity, int level)
        : base(id, name, capacity)
    {
        AutomationLevel = level;
    }

    public override void RemoveItem(Item item)
    {
        Console.WriteLine($"Удаление товара через автоматизированную систему (уровень {AutomationLevel})");
        base.RemoveItem(item);
    }
}

public class StorageBox<T>
{
    public T StoredItem { get; set; }

    public StorageBox(T item)
    {
        StoredItem = item;
    }
}


        Item laptop = new Item(1, "Ноутбук", 0.4);
        Item phone = new Item(2, "Смартфон", 0.1);
        Item server = new Item(3, "Сервер", 3.0);

        Inventory inv1 = new PersonalInventory(101, "Личный склад", 10, "Иван");
        Inventory inv2 = new GroupInventory(102, "Склад электроники", 15, "Электроника");
        Inventory inv3 = new AutomatedInventory(103, "Авто-склад", 20, 3);

        inv1.AddItem(laptop);
        inv2.AddItem(phone, 3);
        inv3.AddItem(server);

        Console.WriteLine();
        inv1.GetStorageStatus();
        inv2.GetStorageStatus();
        inv3.GetStorageStatus();

        Console.WriteLine();

        inv3.RemoveItem(server);

        StorageBox<Item> box = new StorageBox<Item>(phone);
        Console.WriteLine($"\nGeneric StorageBox содержит: {box.StoredItem}");

        Console.WriteLine("\n=== Демонстрация завершена ===");
```

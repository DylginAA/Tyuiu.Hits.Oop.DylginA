```
public class Inventory
{
    public string Name { get; set; }
    public double Capacity { get; set; }
    protected double UsedSpace { get; set; }
    protected List<string> Items { get; set; }
    
 
    public string Location { get; set; }
    public DateTime LastInventoryDate { get; set; }
    public bool IsTemperatureControlled { get; set; }
    public decimal MaintenanceCost { get; set; }

    public Inventory(string name, double capacity, string location)
    {
        Name = name;
        Capacity = capacity;
        Location = location;
        UsedSpace = 0;
        Items = new List<string>();
        LastInventoryDate = DateTime.Now;
        IsTemperatureControlled = false;
        MaintenanceCost = 0;
    }

    public virtual void ShowStatus()
    {
        Console.WriteLine($"{Name} ({Location}): занято {UsedSpace}/{Capacity} м³");
    }

    public virtual void AddItem(string item, double size)
    {
        if (UsedSpace + size <= Capacity)
        {
            Items.Add(item);
            UsedSpace += size;
            Console.WriteLine($"Добавлен '{item}' на {Name}");
        }
        else
        {
            Console.WriteLine($"Не хватает места для '{item}' на {Name}");
        }
    }

    public virtual void RemoveItem(string item, double size)
    {
        if (Items.Remove(item))
        {
            UsedSpace -= size;
            Console.WriteLine($"Удален '{item}' с {Name}");
        }
        else
        {
            Console.WriteLine($"'{item}' не найден на {Name}");
        }
    }
    

    public virtual void PerformMaintenance()
    {
        Console.WriteLine($"Выполняется техническое обслуживание склада {Name}");
        LastInventoryDate = DateTime.Now;
    }
    
    public double GetFreeSpace()
    {
        return Capacity - UsedSpace;
    }
    
    public void ShowItems()
    {
        Console.WriteLine($"Предметы на {Name}: {string.Join(", ", Items)}");
    }
    
    public virtual decimal CalculateMonthlyCost()
    {
        return MaintenanceCost;
    }
}


public class Inventory1 : Inventory
{
    public string Owner { get; set; }
    
   
    public string SecurityLevel { get; set; }
    public bool HasInsurance { get; set; }

    public Inventory1(double capacity, string owner, string location) 
        : base("Склад 1", capacity, location)
    {
        Owner = owner;
        SecurityLevel = "Standard";
        HasInsurance = true;
    }

    public override void ShowStatus()
    {
        Console.WriteLine($"{Name} (Владелец: {Owner}, Безопасность: {SecurityLevel}): занято {UsedSpace}/{Capacity} м³");
    }
    

    public void UpdateSecurity(string newLevel)
    {
        SecurityLevel = newLevel;
        Console.WriteLine($"Уровень безопасности склада {Name} изменен на {newLevel}");
    }
    
    public void ProcessInsuranceClaim()
    {
        if (HasInsurance)
        {
            Console.WriteLine($"Обработка страхового случая для склада {Name}");
        }
    }
}


public class Inventory2 : Inventory
{
    public string Category { get; set; }
    
    
    public int MaxItemWeight { get; set; }
    public bool RequiresSpecialHandling { get; set; }

    public Inventory2(double capacity, string category, string location) 
        : base("Склад 2", capacity, location)
    {
        Category = category;
        MaxItemWeight = 100;
        RequiresSpecialHandling = false;
    }

    public override void AddItem(string item, double size)
    {
        if (UsedSpace + size <= Capacity)
        {
            Items.Add(item);
            UsedSpace += size;
            Console.WriteLine($"Добавлен '{item}' ({Category}) на {Name}");
        }
        else
        {
            Console.WriteLine($"Не хватает места для '{item}' на {Name}");
        }
    }
    
    public bool CanHandleWeight(int weight)
    {
        return weight <= MaxItemWeight;
    }
    
    public void SetSpecialHandling(bool required)
    {
        RequiresSpecialHandling = required;
        Console.WriteLine($"Специальная обработка для {Name}: {(required ? "требуется" : "не требуется")}");
    }
}


public interface IAutomatedSystem
{
    void EnableAutomation();
    void DisableAutomation();
    bool IsSystemOperational();
}

public interface IEnergyEfficient
{
    decimal CalculateEnergyConsumption();
    void EnableEnergySaving();
}


public class Inventory3 : Inventory, IAutomatedSystem, IEnergyEfficient
{
    public int Level { get; set; }
    
    
    public bool IsAutomated { get; set; }
    public double EnergyConsumption { get; set; }
    public int RobotCount { get; set; }

    public Inventory3(double capacity, int level, string location) 
        : base("Склад 3", capacity, location)
    {
        Level = level;
        IsAutomated = false;
        EnergyConsumption = 100.0;
        RobotCount = 0;
    }

    public override void RemoveItem(string item, double size)
    {
        if (Items.Remove(item))
        {
            UsedSpace -= size;
            Console.WriteLine($"Удален '{item}' (Уровень: {Level}) с {Name}");
        }
        else
        {
            Console.WriteLine($"'{item}' не найден на {Name}");
        }
    }
    
  
    public void DeployRobots(int count)
    {
        RobotCount = count;
        Console.WriteLine($"Развернуто {count} роботов на складе {Name}");
    }
    
    
    public void EnableAutomation()
    {
        IsAutomated = true;
        Console.WriteLine($"Автоматизация включена на складе {Name}");
    }
    
    public void DisableAutomation()
    {
        IsAutomated = false;
        Console.WriteLine($"Автоматизация отключена на складе {Name}");
    }
    
    public bool IsSystemOperational()
    {
        return IsAutomated && RobotCount > 0;
    }
    

    public decimal CalculateEnergyConsumption()
    {
        return (decimal)EnergyConsumption * (IsAutomated ? 1.2m : 0.8m);
    }
    
    public void EnableEnergySaving()
    {
        EnergyConsumption *= 0.7;
        Console.WriteLine($"Режим энергосбережения включен на складе {Name}");
    }
    
    public override decimal CalculateMonthlyCost()
    {
        return base.CalculateMonthlyCost() + (decimal)EnergyConsumption * 0.15m;
    }
}


public class SpecializedWarehouse : Inventory, IAutomatedSystem, IEnergyEfficient
{
    public string Specialization { get; set; }
    public bool IsAutomated { get; set; }
    public double EnergyConsumption { get; set; }
    public int SensorCount { get; set; }
    
    
    public string Certification { get; set; }
    public int EmergencyExits { get; set; }

    public SpecializedWarehouse(string name, double capacity, string location, string specialization) 
        : base(name, capacity, location)
    {
        Specialization = specialization;
        IsAutomated = true;
        EnergyConsumption = 150.0;
        SensorCount = 10;
        Certification = "ISO-9001";
        EmergencyExits = 4;
    }

  
    public void ConductInspection()
    {
        Console.WriteLine($"Проведена проверка сертификации {Certification} для {Name}");
    }
    
    public void UpdateCertification(string newCertification)
    {
        Certification = newCertification;
        Console.WriteLine($"Сертификация обновлена на {newCertification} для {Name}");
    }
    
 
    public void EnableAutomation()
    {
        IsAutomated = true;
        Console.WriteLine($"Автоматизация включена на специализированном складе {Name}");
    }
    
    public void DisableAutomation()
    {
        IsAutomated = false;
        Console.WriteLine($"Автоматизация отключена на специализированном складе {Name}");
    }
    
    public bool IsSystemOperational()
    {
        return IsAutomated && SensorCount > 5;
    }
    

    public decimal CalculateEnergyConsumption()
    {
        return (decimal)EnergyConsumption * (IsAutomated ? 1.1m : 0.9m);
    }
    
    public void EnableEnergySaving()
    {
        EnergyConsumption *= 0.6;
        Console.WriteLine($"Режим энергосбережения включен на специализированном складе {Name}");
    }
    
    public override void PerformMaintenance()
    {
        base.PerformMaintenance();
        Console.WriteLine($"Дополнительная проверка систем безопасности на {Name}");
    }
}

        Console.WriteLine("=== Расширенная система складов ===\n");

        // Простое наследование
        Inventory1 sklad1 = new Inventory1(100, "Иванов", "Москва");
        Inventory2 sklad2 = new Inventory2(80, "Электроника", "Санкт-Петербург");
        
        // Сложное наследование
        Inventory3 sklad3 = new Inventory3(120, 3, "Казань");
        
        // Множественное наследование
        SpecializedWarehouse specSklad = new SpecializedWarehouse("ХимСклад", 200, "Новосибирск", "Химические материалы");

        Console.WriteLine("1. Добавление предметов:");
        sklad1.AddItem("Стол", 10);
        sklad1.AddItem("Стул", 5);
        
        sklad2.AddItem("Ноутбук", 3);
        sklad2.AddItem("Телефон", 1);
        
        sklad3.AddItem("Принтер", 8);
        sklad3.AddItem("Монитор", 12);
        
        specSklad.AddItem("Реактивы", 15);
        specSklad.AddItem("Оборудование", 25);

        Console.WriteLine("\n2. Статус складов:");
        sklad1.ShowStatus();
        sklad2.ShowStatus();
        sklad3.ShowStatus();
        specSklad.ShowStatus();

        Console.WriteLine("\n3. Использование новых методов:");
        sklad1.UpdateSecurity("High");
        sklad1.ProcessInsuranceClaim();
        
        Console.WriteLine($"Может ли склад 2 обработать вес 150кг: {sklad2.CanHandleWeight(150)}");
        sklad2.SetSpecialHandling(true);
        
        sklad3.EnableAutomation();
        sklad3.DeployRobots(5);
        Console.WriteLine($"Система склада 3 оперативна: {sklad3.IsSystemOperational()}");
        Console.WriteLine($"Потребление энергии: {sklad3.CalculateEnergyConsumption()} кВт");
        
        specSklad.ConductInspection();
        specSklad.UpdateCertification("ISO-14001");
        specSklad.EnableEnergySaving();

        Console.WriteLine("\n4. Удаление предметов:");
        sklad1.RemoveItem("Стул", 5);
        sklad3.RemoveItem("Принтер", 8);

        Console.WriteLine("\n5. Техническое обслуживание:");
        sklad1.PerformMaintenance();
        sklad3.PerformMaintenance();
        specSklad.PerformMaintenance();

        Console.WriteLine("\n6. Финансовые расчеты:");
        Console.WriteLine($"Месячная стоимость склада 1: {sklad1.CalculateMonthlyCost()} руб.");
        Console.WriteLine($"Месячная стоимость склада 3: {sklad3.CalculateMonthlyCost()} руб.");
        Console.WriteLine($"Месячная стоимость специализированного склада: {specSklad.CalculateMonthlyCost()} руб.");

        Console.WriteLine("\n7. Свободное пространство:");
        Console.WriteLine($"Свободно на складе 1: {sklad1.GetFreeSpace()} м³");
        Console.WriteLine($"Свободно на складе 2: {sklad2.GetFreeSpace()} м³");

        Console.WriteLine("\n8. Список предметов:");
        sklad1.ShowItems();
        sklad2.ShowItems();
        sklad3.ShowItems();
        specSklad.ShowItems();
        Console.WriteLine("\n9. Демонстрация полиморфизма:");
        Inventory[] allWarehouses = { sklad1, sklad2, sklad3, specSklad };
        
        foreach (var warehouse in allWarehouses)
        {
            warehouse.ShowStatus();
            warehouse.PerformMaintenance();
            
            if (warehouse is IAutomatedSystem automated)
            {
                Console.WriteLine($"Автоматизированная система: {automated.IsSystemOperational()}");
            }
            
            if (warehouse is IEnergyEfficient energyEfficient)
            {
                Console.WriteLine($"Потребление энергии: {energyEfficient.CalculateEnergyConsumption()} кВт");
            }
            
            Console.WriteLine("---");
        }
```
